```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from keras.callbacks import ModelCheckpoint
from sklearn import preprocessing
from numpy import argmax 
```

```python
def get_dataset():
    #get train data
    dataset_path ='drive/MyDrive/TIES4911/demo1/Automobile_price_data_Raw_set.csv'
    dataset = pd.read_csv(dataset_path)
    
    return dataset

def get_combined_data():
  #reading train data
  dataset = get_dataset()
  dataset.dropna(axis=0, inplace=True)

  target = dataset.make
  dataset.drop(['make'],axis = 1 , inplace = True)

  combined = dataset
  combined.reset_index(inplace=True)
  combined = dataset[['body-style', 'wheel-base', 'engine-size', 'horsepower', 'peak-rpm', 'highway-mpg', 'price']]
  return combined, target
```

```python
dataset = get_dataset()
combined, target = get_combined_data()
```

```python
def get_cols_with_no_nans(df,col_type):
    if (col_type == 'num'):
        predictors = df.select_dtypes(exclude=['object'])
    elif (col_type == 'no_num'):
        predictors = df.select_dtypes(include=['object'])
    elif (col_type == 'all'):
        predictors = df
    else :
        print('Error : choose a type (num, no_num, all)')
        return 0
    cols_with_no_nans = []
    for col in predictors.columns:
        if not df[col].isnull().any():
            cols_with_no_nans.append(col)
    return cols_with_no_nans
```

```python
num_cols = get_cols_with_no_nans(combined , 'num')
cat_cols = get_cols_with_no_nans(combined , 'no_num')
```

```python
print ('Number of numerical columns with no nan values :',len(num_cols))
print ('Number of nun-numerical columns with no nan values :',len(cat_cols))
```

```python
def oneHotEncode(df,colNames):
    for col in colNames:
        if( df[col].dtype == np.dtype('object')):
            dummies = pd.get_dummies(df[col],prefix=col)
            df = pd.concat([df,dummies],axis=1)

            #drop the encoded column
            df.drop([col],axis = 1 , inplace=True)
    return df
    

print('There were {} columns before encoding categorical features'.format(combined.shape[1]))
combined = oneHotEncode(combined, cat_cols)
print('There are {} columns after encoding categorical features'.format(combined.shape[1]))
```

```python
label_encoder = preprocessing.LabelEncoder()
labels = label_encoder.fit(target)
```

```python
num_labels = len(list(labels.classes_))
```

```python
labels_encoded = label_encoder.fit_transform(target)
```

```python
labels_encoded
```

```python
x_train, x_test, y_train, y_test = train_test_split(combined, labels_encoded, test_size = 0.2,random_state = 0)
```

```python
NN_model = Sequential()

# The Input Layer :
NN_model.add(Dense(128, kernel_initializer='normal',input_dim = combined.shape[1], activation='relu'))

# The Hidden Layers :
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))

# The Output Layer :
NN_model.add(Dense(num_labels, activation='softmax'))

# Compile the network :
NN_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
NN_model.summary()
```

```python
NN_model.fit(x_train, y_train, epochs=50, batch_size=32, validation_split = 0.2)
```

```python
NN_model.evaluate(x_test, y_test)
```
