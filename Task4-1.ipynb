```python
import tensorflow as tf
from tensorflow.keras.applications import resnet50, xception, inception_v3  
from tensorflow.keras.preprocessing import image
import numpy as np
from keras import Input, Model
import matplotlib.pyplot as plt
```

```python
img_path = 'drive/My Drive/TIES4911/task4/eggnog.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img = image.img_to_array(img)
```

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip('horizontal'),
  tf.keras.layers.RandomRotation(0.2),
])
```

```python
input_tensor = Input(shape=(224, 224, 3))

resnet50_model = resnet50.ResNet50()
xception_model = xception.Xception(input_tensor=input_tensor)
inceptionv3_model = inception_v3.InceptionV3(input_tensor=input_tensor)
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
  img = data_augmentation(img)

  ax = plt.subplot(1, 3, 1)
  plt.imshow(img.numpy().astype("uint8"))
  plt.title('Augmentation {}'.format(i + 1))
  plt.axis("off")
  plt.show()

  x = np.expand_dims(img, axis=0)
  x = x.copy()

  x = resnet50.preprocess_input(x)
  preds = resnet50_model.predict(x, verbose=0)
  print('ResNet50 preds: ', resnet50.decode_predictions(preds, top=3))

  x = xception.preprocess_input(x)
  preds = xception_model.predict(x, verbose=0)
  print('xception preds: ', xception.decode_predictions(preds, top=3))

  x = inception_v3.preprocess_input(x)
  preds = inceptionv3_model.predict(x, verbose=0)
  print('inception_v3 preds: ', inception_v3.decode_predictions(preds, top=3))
  print()
```
